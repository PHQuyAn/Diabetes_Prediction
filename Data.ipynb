```python
import pandas as pd
import re

# Đọc file Excel
file_path = "/content/Bệnh + Mã.xlsx"  # Thay đổi đường dẫn đến file Excel của bạn
df = pd.read_excel(file_path)

# Chuẩn hóa tên cột để tránh lỗi chính tả hoặc khoảng trắng
df.columns = df.columns.str.strip().str.lower()

# In danh sách các cột để kiểm tra
print("Các cột trong file Excel:", df.columns)

# Tìm cột gần đúng nếu có sai khác
expected_column_name = "danh sách bệnh"
matched_columns = [col for col in df.columns if expected_column_name in col]

if matched_columns:
    column_name = matched_columns[0]  # Chọn cột khớp gần nhất
else:
    raise KeyError(f"Cột '{expected_column_name}' không tồn tại trong file Excel. Hãy kiểm tra lại!")

# Trích xuất dữ liệu của cột "Danh sách bệnh"
diagnosis_column = df[column_name].astype(str)  # Đảm bảo dữ liệu dạng chuỗi

# Hàm phân loại chẩn đoán thành danh sách bệnh (mã bệnh)
def extract_diagnosis_list(diagnosis):
    matches = re.findall(r'([^()]+)\(([^)]+)\)', diagnosis)  # Tìm bệnh theo format "Tên bệnh (Mã)"
    return [f"{match[0].strip()} ({match[1].strip()})" for match in matches] if matches else []

# Áp dụng hàm phân loại
df[column_name] = diagnosis_column.apply(extract_diagnosis_list)

# Chuyển danh sách bệnh thành dạng một dòng mỗi bệnh
df_exploded = df.explode(column_name)

# Danh sách từ khóa phân loại bệnh
diabetes_keywords = ["tiểu đường", "đái tháo đường", "glucose", "insulin", "HBA1c", "tăng đường huyết", "giảm đường huyết"]
cardiovascular_keywords = ["tăng huyết áp", "bệnh mạch vành", "suy tim", "nhồi máu cơ tim", "đột quỵ"]
liver_keywords = ["viêm gan", "xơ gan", "gan nhiễm mỡ"]
kidney_keywords = ["suy thận", "hội chứng thận hư", "viêm cầu thận", "sỏi thận"]
digestive_keywords = ["viêm loét dạ dày", "hội chứng ruột kích thích", "viêm ruột", "khó tiêu"]
orthopedic_keywords = ["thoái hóa khớp", "loãng xương", "gãy xương", "viêm khớp", "đau lưng", "viêm cột sống"]
ear_keywords = ["viêm tai", "ù tai", "điếc", "viêm màng nhĩ", "rối loạn tiền đình"]
blood_keywords = ["thiếu máu", "rối loạn đông máu", "bạch cầu", "xuất huyết", "tăng lipid máu"]
spine_keywords = ["thoái hóa cột sống", "gai cột sống", "chấn thương cột sống", "đau lưng"]
skull_keywords = ["chấn thương sọ não", "tổn thương hộp sọ", "xuất huyết não", "tổn thương nội sọ", "tổn thương nông ở đầu"]

# Hàm phân loại bệnh chi tiết
def classify_disease(disease):
    if not isinstance(disease, str):
        return "Bệnh khác"
    disease_lower = disease.lower()
    if any(keyword in disease_lower for keyword in diabetes_keywords):
        return "Bệnh tiểu đường"
    elif any(keyword in disease_lower for keyword in cardiovascular_keywords):
        return "Bệnh tim mạch"
    elif any(keyword in disease_lower for keyword in liver_keywords):
        return "Bệnh gan"
    elif any(keyword in disease_lower for keyword in kidney_keywords):
        return "Bệnh thận"
    elif any(keyword in disease_lower for keyword in digestive_keywords):
        return "Bệnh tiêu hóa"
    elif any(keyword in disease_lower for keyword in orthopedic_keywords):
        return "Bệnh xương khớp"
    elif any(keyword in disease_lower for keyword in ear_keywords):
        return "Bệnh về tai"
    elif any(keyword in disease_lower for keyword in blood_keywords):
        return "Bệnh về máu"
    elif any(keyword in disease_lower for keyword in spine_keywords):
        return "Bệnh cột sống"
    elif any(keyword in disease_lower for keyword in skull_keywords):
        return "Bệnh hộp sọ"
    else:
        return "Bệnh khác"

# Áp dụng phân loại
df_exploded["Nhóm bệnh"] = df_exploded[column_name].apply(classify_disease)

# Xuất file Excel
output_file = "/content/XN_Processed_Output.xlsx"
df_exploded.to_excel(output_file, index=False)
print(f"File đã được lưu tại: {output_file}")
```
